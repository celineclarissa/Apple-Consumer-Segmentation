## Abstract
Apple Inc.’s marketing strategies are reported to impact university students’ consumption behavior. However, previous studies focused on overall students, not specifically on international students in China. This demographic has a more equal exposure to both cultures. Quantitative and qualitative approaches were used to study 52 international students at a Sino-foreign university, utilizing surveys, proceeded with data clustering and statistical analyses. Results show that within the demographic, there were two groups with different perceptions towards Apple products: skeptic and enthusiast, directly influencing purchase intentions. Findings align with previous research about university students’ and society’s view on Apple products.

## Introduction
Apple Inc. is a world-renowned company specializing in technological products such as iPhones, iPads, MacBooks, to AirPods and other electrical accessories. Their choice of design style in goods and their unique brand presence act as an appeal to customers. This aligns with Apple Inc.’s core branding strategy: empathy, focus, impute, which are reflected in how they utilize humanity, building emotional bonds with users and affecting consumer behavior (Cai & Chen, 2024).

China, with a population of 1.4 billion (World Bank, 2024), acts as a prominent market for Apple Inc.’s brand expansion. This includes Shenzhen, an economic and technological hub, and a rising immigrant city (Information Office of Shenzhen Municipal Government). One of the institutions attracting international students is The Chinese University of Hong Kong, Shenzhen, a university promoting cultural exchange among their students. International students at The Chinese University of Hong Kong, Shenzhen are exposed to both western and eastern media, culture, and technologies. Past studies have examined about Apple Inc.’s marketing strategy in China and other countries, but little is known about how it affects the perception and behavior of international students who have a more equal access to both environments. Therefore, this research aims to bridge that gap, providing information for both companies to further assess current initiatives towards this market segment and the people.

## Literature Review
Apple Inc.’s growth is a result of their innovations, design, cutting-edge operating system, and cloud computing technology (Chen et al., 2021). As a global leading company, not only their technologies are exceptional, but also their unique branding among other enterprises in the industry. This initiative includes framing their products as a social symbol and enhancing consumer experience in retail stores (Lan, 2021). This strategy targets key factors of consumer behavior for their brand, which Nagamia (2015) identified as psychological aspects and sociocultural aspects, such as lifestyle and social status. 

Douglas (2021) found that Apple products are romanticized by undergraduate students in Nigeria, showing how lifestyle and symbolic value contributes to purchasing decisions. Additionally, Li (2023) found that Apple Inc.’s brand positioning and marketing strategy is effective in attaining a positive reputation within the Chinese market. However, there was a significant gap in understanding how Apple Inc.’s marketing schemes are interpreted by those the younger generation who are exposed to both Chinese and western media, culture, and technologies, which according to Wells & Prensky (1996) can affect consumer behavior. This demographic included international students studying at The Chinese University of Hong Kong, Shenzhen. Hence, this research focused on answering the research questions below.
1.	How do international students at The Chinese University of Hong Kong, Shenzhen perceive Apple Inc.’s brand identity?
2.	What factors drive or hinder international students at The Chinese University of Hong Kong, Shenzhen from purchasing Apple products?

## Methods
A Google Form, including Likert-scales, checkboxes, multiple-choice, and open-ended questions was used to collect quantitative and qualitative data from 52 active CUHK-Shenzhen international students to gain a comprehensive understanding of the research problem. The questionnaire included three sections: demographic, brand perception, purchase drivers and barriers.

The data was used for quantitative and qualitative analyses. K-means clustering was used to group respondents based on their perception and behavioral drivers (psychographic and behavioral-based consumer segmentation), an important for marketing strategies (Schlegelmilch, 2022). Then, descriptive and inferential statistics (mean and two-sample t-test) were used to identify trends, while the qualitative analysis utilized a thematic method to enhance findings from the quantitative analyses.

#### Model Definition

```python
sns.scatterplot(x=df_scaled[:,0], y=df_scaled[:,1])
```

```python
wcss = []
random_state = 10
max_cluster = 5
for i in range(2, max_cluster+1):
    km = KMeans(n_clusters = i,
                init = 'k-means++',
                max_iter = 300,
                n_init = 10,
                random_state = random_state)
    km.fit(df_scaled)
    wcss.append(km.inertia_)

plt.plot(range(2, max_cluster+1), wcss, marker ="o")
plt.grid()
plt.title('Elbow Method', fontsize = 20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
def plot_silhouette(range_n_clusters, X, random_state):
    for n_clusters in range_n_clusters:
        # Create a subplot with 1 row and 2 columns
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.set_size_inches(18, 4)

        # The 1st subplot is the silhouette plot
        # The silhouette coefficient can range from -1, 1 but in this example all lie within [-0.1, 1]
        ax1.set_xlim([-0.1, 1])
        # The (n_clusters+1)*10 is for inserting blank space between silhouette plots of individual clusters, to demarcate them clearly.
        ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

        # Initialize the clusterer with n_clusters value and a random generator seed of 10 for reproducibility.
        clusterer = KMeans(n_clusters = n_clusters, init = 'k-means++', max_iter = 300, n_init = 10, random_state = random_state)
        cluster_labels = clusterer.fit_predict(X)

        # The silhouette_score gives the average value for all the samples.
        # This gives a perspective into the density and separation of the formed clusters
        silhouette_avg = silhouette_score(X, cluster_labels)
        print("For n_clusters =", n_clusters,
              "The average silhouette_score is :", silhouette_avg)

        # Compute the silhouette scores for each sample
        sample_silhouette_values = silhouette_samples(X, cluster_labels)

        y_lower = 10
        for i in range(n_clusters):
            # Aggregate the silhouette scores for samples belonging to cluster i, and sort them
            ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_clusters)
            ax1.fill_betweenx(np.arange(y_lower, y_upper),
                              0, ith_cluster_silhouette_values,
                              facecolor=color, edgecolor=color, alpha=0.7)

            # Label the silhouette plots with their cluster numbers at the middle
            ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

            # Compute the new y_lower for next plot
            y_lower = y_upper + 10  # 10 for the 0 samples

        ax1.set_title("The silhouette plot for the various clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")

        # The vertical line for average silhouette score of all the values
        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

        ax1.set_yticks([])  # Clear the yaxis labels / ticks
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

        # 2nd Plot showing the actual clusters formed
        colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
        ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                    c=colors, edgecolor='k')

        # Labeling the clusters
        centers = clusterer.cluster_centers_

        # Draw white circles at cluster centers
        ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                    c="white", alpha=1, s=200, edgecolor='k')

        for i, c in enumerate(centers):
            ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50, edgecolor='k')

        ax2.set_title("The visualization of the clustered data.")
        ax2.set_xlabel("Feature space for the 1st feature")
        ax2.set_ylabel("Feature space for the 2nd feature")

        plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                      "with n_clusters = %d" % n_clusters),
                     fontsize=14, fontweight='bold')
```

```python
plot_silhouette(range(2, max_cluster), df_scaled, random_state)
```

## Discussion
The characteristics of the two groups can be identified from the results. Group 1 gave lower ratings with mixed and negative descriptors of the brand and only highly prioritize fundamental factors (price and technical specifications) when purchasing a technological product, making them less prone to symbolic branding. Contrastingly, group 2 gave higher ratings with positive reviews and focused on overall user experience by ecosystem, operating system, and design on top of the fundamental considerations. Additionally, group 1 owns less Apple products compared to group 2. This shows that the students’ views and key considerations of purchase directly impacts their decision. 

Group 1 holds a more skeptical position towards Apple products because their highest priority are only the fundamental factors, with lower levels of considerations for other factors, making them less likely to be swayed by marketing strategies that puts user experience and exclusivity as a driver of purchase, which is Apple’s current strategy. Moreover, this strategy seems to backfire for group 1, looking at their choices of words including “overrated” and “overpriced” to describe Apple. In contrast, group 2 is Apple’s targeted market. On top of the fundamental factors, they also consider product design, operating system, and ecosystem, which are secondary to functionality but enhances their experience while using the products, affecting their purchasing decisions. They also described Apple as “exclusive” and “premium”, indicating Apple’s success in their branding strategy for this market segment. Schlegelmilch (2022) confirmed that segmenting, targeting, and positioning are crucial elements in marketing strategies. Nagamia (2015) confirms the use of this strategy by Apple by mentioning that Apple targeting customers with higher levels of lifestyles.

Results also show that regardless of grouping and ownership of Apple products, the respondents highly consider pricing. The two groups’ purchase decisions differ despite the similarities in this aspect. Group 1 perceived Apple as “overpriced”, showing a negative view towards Apple products’ pricing. Meanwhile, group 2 describing Apple as “exclusive” shows that Apple’s high pricing enhance their positive view of Apple. This is supported by findings showing that price sensitivity only has a moderate influence toward purchase intentions (Shang & Plaikaew, 2025). The same research indicated that some market segments perceive high prices as indicators of high quality, the difference in perceptions.

## Conclusion
As a conclusion, within the main market segment of international students at The Chinese University of Hong Kong, Shenzhen, there were two sub-segments with noticeable differences in their views and behavior towards Apple. One group was more skeptical and emphasized only functionality and price, while the other was more enthusiastic and considered user experience. These beliefs shaped their purchasing decisions. Despite the differences, however, all respondents emphasize pricing in purchasing decisions.

This research lacks depth in qualitative understanding of the respondents’ experience such as the reason why the Apple ecosystem eases daily usage of electronics as a student. Hence, further qualitative methods are encouraged for future research to gain knowledge about the respondents. 

This research implies that symbolic-based branding schemes alone are not sufficient for this specific market segment. Furthermore, companies should consider the dynamics and sub-segments in each target market, then adjust marketing strategies accordingly.
